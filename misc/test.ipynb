```python
import re
date_string = "Ngày 02 tháng 10 năm2019  Since2013"
def get_num_from_date_string_by_prefix(date_string_, prefix_):
    prefix = prefix_.lower()
    date_string = date_string_.lower()
    if  prefix in date_string:
        match = re.split(fr"{prefix}\s*(\d+\s*\d*)\s*\b", date_string)
        num = match[1] 
        remain_string = match[2] if prefix!="năm" else match[0]
        return num, remain_string
    else:
        return "xx", date_string_ if prefix != "năm" else "xxxx", date_string_

year, remain_str = get_num_from_date_string_by_prefix(date_string, "năm")
print(year, remain_str)
day, remain_str = get_num_from_date_string_by_prefix(remain_str, "ngày")
print(day, remain_str)
month, remain_str = get_num_from_date_string_by_prefix(remain_str, "tháng")
print(month, remain_str)
```
